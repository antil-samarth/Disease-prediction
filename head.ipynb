# Diseases List

working on release_conditions.json

```python
import json
```

```python
with open('Data/release_conditions.json') as f:
    diseases = json.load(f)
    
diseases
```

```python
diseases_list = list(diseases.keys())

diseases_dict = {}

for key, value in diseases.items():
    value['symptoms'].update(value['antecedents'])
    
for key, value in diseases.items():
    diseases_dict[key] = value['symptoms']

for key, value in diseases_dict.items():
    diseases_dict[key] = list(value.keys())
    
```
