```python
with open('Data/release_evidences.json') as f:
    symptom_questions = json.load(f)
    
c=0    
for z in symptom_questions:
    print(symptom_questions[z])
    c+=1
    if c==5:
        break
```

```python
symptom_questions_dict = {}

for key, value in symptom_questions.items():
    symptom_questions_dict[key] = value['question_en']

len(symptom_questions_dict)
```

```python
symptom_questions_list = list(symptom_questions_dict.keys())
symptom_questions_list.sort()

c=0
for key, value in symptom_questions_dict.items():
    print(key, value)
    c+=1
    if c==5:
        break
```

```python
symptom_response_type = {}
response_type = ""

for key, value in symptom_questions.items():
    if value['data_type'] == 'B':
        response_type = 'bool'
    elif value['data_type'] == 'C':
        response_type = 'str'
    else:
        response_type = 'list'
        
    symptom_response_type[key] = response_type
    
#count of each response type
from collections import Counter 
print(Counter(symptom_response_type.values()))
```

```python
evi = df['EVIDENCES']
evi = evi.copy()
evi.head()
```

```python
trial = data.copy()
trial.head()
```

```python
cols_to_zero = [col for col in trial.columns if col not in ['age', 'sex']]
trial[cols_to_zero] = 0
```

```python
trial.head()
```

```python
for i in range(len(evi)):
    evi[i] = evi[i][1:-1].split(', ')
```

```python
evi.head()
```

```python
for i in range(len(evi)):
    for j in range(len(evi[i])):
        #print(i, j, evi[i][j])
        evi[i][j] = evi[i][j].replace("'", '')
        #print(evi[i][j], type(evi[i][j]))
        
        if evi[i][j] not in symptom_questions_list:
            if evi[i][j].split('_@_')[1] not in null_response:    
                trial.loc[i, evi[i][j].split('_@_')[0]] = 1
        else:
            trial.loc[i, evi[i][j]] = 1
```

```python
trial = trial.copy()
trial['pathology'] = target
```
