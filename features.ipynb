```python
import pandas as pd
import numpy as np
```

```python
# Load the data

df = pd.read_csv('Data/processed/processed.csv')
```

```python
df.head()
```

```python
#calulate the percentage of each pathology

df['pathology'].value_counts(normalize=True)
```

```python
# drop age, sex and pathology columns

df = df.drop(['age', 'sex', 'pathology'], axis=1)
```

```python
df.head()
```

```python
# Calculate density
density = df.mean(axis=0).mean()
density
```

```python
# Calculate support
support = (df != 0).any(axis=0).sum()
support
```

```python
# Calculate sparsity level per feature
sparsity_per_feature = (1 - df.mean(axis=0)) * 100
print(f"Sparsity level per feature:\n{sparsity_per_feature.to_string()}")
```

```python
corr_matrix = df.corr(method='spearman')
```

```python
threshold = 0.8

high_corr = []
for i in range(len(corr_matrix)):
    for j in range(i + 1, len(corr_matrix)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_corr.append((corr_matrix.index[i], corr_matrix.index[j], corr_matrix.iloc[i, j]))

print("Features with correlation equal to", threshold, ":")
for feature1, feature2, correlation in high_corr:
    print(f"{feature1}, {feature2} - Correlation: {correlation:.4f}")
```
