```python
import pandas as pd
data = pd.read_csv('Data/processed/processed.csv')
```

```python
#calulate the percentage of each pathology

pct_dist = data['pathology'].value_counts(normalize=True)
pct_dist = pct_dist * 100
pct_dist
```

```python
ax = pct_dist.plot(kind='barh', 
                   title='Percentage distribution of pathologies', 
                   figsize=(8, 12), 
                   color='skyblue')
```

```python
#seperate pathology, age and sex from data

pathology = data['pathology']
age = data['age']
sex = data['sex']

data = data.drop(['pathology', 'age', 'sex'], axis=1)
```

```python
data.head()
```

```python
# Calculate density
density = data.mean(axis=0).mean()
density
```

```python
# Calculate support
support = (data != 0).any(axis=0).sum()
support
```

```python
# Calculate sparsity level per feature
sparsity_per_feature = (1 - data.mean(axis=0)) * 100
print(f"Sparsity level per feature:\n{sparsity_per_feature.to_string()}")
```

```python
corr_matrix = data.corr(method='spearman')
```

```python
threshold = 0.8

high_corr = []
for i in range(len(corr_matrix)):
    for j in range(i + 1, len(corr_matrix)):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            high_corr.append((corr_matrix.index[i], corr_matrix.index[j], corr_matrix.iloc[i, j]))

print("Features with correlation equal to", threshold, ":")
for feature1, feature2, correlation in high_corr:
    print(f"{feature1}, {feature2} - Correlation: {correlation:.4f}")
```
